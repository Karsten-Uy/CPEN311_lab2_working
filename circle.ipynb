```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['scatter.marker'] = 's'

def cust_scatter(x, y, **kwargs):
    kwargs.setdefault('marker', 's')  # default square
    kwargs.setdefault('s', 10)        # default size
    return plt.scatter(x, y, **kwargs)

def drawGreenSegment(centre_x, centre_y, radius, offset_x, offset_y):
    if (centre_x + offset_x >= c_x3):
        cust_scatter(centre_x + offset_x, centre_y - offset_y, c="green", alpha=0.3) # -- octant 8
    if (centre_x + offset_y >= c_x3):
        cust_scatter(centre_x + offset_y, centre_y - offset_x, c="green", alpha=0.3) # -- octant 7

def drawBlueSegment(centre_x, centre_y, radius, offset_x, offset_y):
    if (centre_x - offset_x <= c_x3):
        cust_scatter(centre_x - offset_x, centre_y - offset_y, c="blue", alpha=0.3) # -- octant 5
    if (centre_x - offset_y <= c_x3):
        cust_scatter(centre_x - offset_y, centre_y - offset_x, c="blue", alpha=0.3) # -- octant 6

def drawRedSegment(centre_x, centre_y, radius, offset_x, offset_y):
    if (centre_x + offset_y <= c_x1 and centre_x + offset_y >= c_x2):
        cust_scatter(centre_x + offset_y, centre_y + offset_x, c="red", alpha=0.3) # -- octant 2
    if (centre_x - offset_y <= c_x1 and centre_x - offset_y >= c_x2):
        cust_scatter(centre_x - offset_y, centre_y + offset_x, c="red", alpha=0.3) # -- octant 3

def draw_triangle_segment(centre_x, centre_y, radius, type):
    
    offset_y = 0
    offset_x = radius
    crit = 1 - radius
    num_points = 0

    while offset_y <= offset_x:
        if (type == "green"):
            drawGreenSegment(centre_x, centre_y, radius, offset_x, offset_y)
        elif (type == "blue"):
            drawBlueSegment(centre_x, centre_y, radius, offset_x, offset_y)
        elif (type == "red"):
            drawRedSegment(centre_x, centre_y, radius, offset_x, offset_y)

        offset_y = offset_y + 1
        if crit <= 0:
            crit = crit + 2 * offset_y + 1
        else:
            offset_x = offset_x - 1
            crit = crit + 2 * (offset_y - offset_x) + 1

    plt.gca().set_aspect("equal")
    # plt.grid()
    # plt.show()

centre_x = 96;
centre_y = 43;
radius   = 64;
diameter = radius;

c_x = centre_x;
c_y = centre_y;
c_x1 = c_x + diameter/2;
c_y1 = c_y + diameter * np.sqrt(3)/6;
c_x2 = c_x - diameter/2;
c_y2 = c_y + diameter * np.sqrt(3)/6;
c_x3 = c_x;
c_y3 = c_y - diameter * np.sqrt(3)/3;

plt.scatter(int(c_x1), int(c_y1), color="red", linewidths=5)
plt.scatter(int(c_x2), int(c_y2), color="red", linewidths=5)
plt.scatter(int(c_x3), int(c_y3), color="red", linewidths=5)

plt.text(int(c_x1) + 2, int(c_y1), "c_x1, c_y1")
plt.text(int(c_x2) - 2, int(c_y2), "c_x2, c_y2")
plt.text(int(c_x3) + 2, int(c_y3), "c_x3, c_y3")

draw_triangle_segment(int(c_x2), int(c_y2), radius, type="green")
draw_triangle_segment(int(c_x1), int(c_y1), radius, type="blue")
draw_triangle_segment(int(c_x3), int(c_y3), radius, type="red")

plt.gca().invert_yaxis()
```
